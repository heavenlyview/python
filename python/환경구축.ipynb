### 문제

- 날씨 정보를 받습니다.  
1. 20220809Sunny
2. 20220810Cloudy
3. 20220811Fog

- 이 정보들을 연도 / 월/ 일 / 날씨 변수로 쪼개서 저장하고 싶어요
- 어떤 날씨 정보가 오든지 간에 연/월/일 날씨를 변함없이 지칭하는 슬라이싱 코드 짜기

```python
weather_info = "20220809Fog"
# weather_info = "20220810Sunny"
# weather_info = "20220811Cloudy"

year = weather_info[0:4]
month = weather_info[4:6]
day = weather_info[6:8]
weather = weather_info[8:]

print(year)
print(month)
print(day)
print(weather)
```
